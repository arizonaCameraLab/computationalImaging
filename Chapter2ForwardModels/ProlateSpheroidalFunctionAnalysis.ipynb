```python
# prompt: plot the prolate spheroidal wavefunctions of order 0 to 10 from x= 0 to 5

import numpy as np
from scipy.special import pro_ang1
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 100)

n_values = np.arange(0, 9)
plt.figure(figsize=(15, 15))
for n in n_values:
  # Calculate the prolate spheroidal wavefunction of order n
  s= pro_ang1(0, n, 10.0, x)  # Example c=1.
  plt.subplot(3, 3, n+1)
  plt.plot(x, s[0], label=f"n={n}")
  plt.title(fr"$\psi_{n}$")

plt.show()
```
