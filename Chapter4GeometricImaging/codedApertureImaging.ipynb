## Circ function

```python
# prompt: make a 3D plot of circ(x,y) over the 2D plane


def circ(x, y, radius):
  z = np.sqrt(x**2 + y**2)
  return(z<=radius)

side = np.linspace(-1,1,1024)
X,Y = np.meshgrid(side,side)
Z = circ(X,Y,0.5)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

### Quadratic residues

```python
def qres(p):
  #function to find the quadratic residues modulo prime number p between 1
  # and p-1. The function returns a list with value 1 for all indices which
  # are quadratic residues modulo p and value 0 for indices which are not
  # quadratic residues.
  #
  qr=np.zeros(p)
  for i in range(1,p):
    qr[np.mod(i**2,p)-1]=1
  return qr
```

# Neural denoising

## load and prepare data

```python
# load the mnist dataset of characters
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train=x_train.astype("float32")/255.0
x_test=x_test.astype("float32")/255.0


g_test=np.zeros(np.shape(x_test))
g_train=np.zeros(np.shape(x_train))
for pop in range(np.shape(x_test)[0]):
  g_test[pop,:,:]=cv2.resize(signal.convolve2d(ht,np.random.poisson(5*signal.convolve2d(t,x_test[pop,:,:], mode='same',boundary='wrap'))\
                                               ,mode='same',boundary='wrap'),dsize=(28,28))
  g_test[pop,:,:]=g_test[pop,:,:]/np.max(g_test[pop,:,:])
for pop in range(np.shape(x_train)[0]):
  g_train[pop,:,:]=cv2.resize(signal.convolve2d(ht,np.random.poisson(5*signal.convolve2d(t,x_train[pop,:,:], mode='same',boundary='wrap'))\
                                                ,mode='same',boundary='wrap'),dsize=(28,28))
  g_train[pop,:,:]=g_train[pop,:,:]/np.max(g_train[pop,:,:])
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1))
x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))
g_train = np.reshape(g_train, (len(g_train), 28, 28, 1))
g_test = np.reshape(g_test, (len(g_test), 28, 28, 1))
```

### Encoder/denoising model

we use a denoising model drawn from https://blog.keras.io/building-autoencoders-in-keras.html

```python
input_img = keras.Input(shape=(28, 28, 1))

x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = layers.MaxPooling2D((2, 2), padding='same')(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
encoded = layers.MaxPooling2D((2, 2), padding='same')(x)

# At this point the representation is (7, 7, 32)

x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(encoded)
x = layers.UpSampling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x)
decoded = layers.Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = keras.Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
autoencoder.summary()
```

### Training

```python

autoencoder.fit(
    x=g_train,
    y=x_train,
    epochs=10,
    batch_size=128,
    shuffle=True,
    validation_data=(g_test, x_test),
)
```

### random signal

In using coded apertures to encode the measurement data we implicitly assume that we can use the spatial structure of measurements to recover the image. If the image is actually a random pattern this assumption may fail and denoising may not be possible. The following simple example looks at reconstruction of a quasi-random pattern in the presence of noise.

```python

imagey=np.zeros((31,31))
imagey[8:24,8:24]=np.random.rand(16,16)
#imagey=x_train[46];
plt.figure(figsize=(15, 7))
plt.rcParams.update({'font.size': 22})
plt.subplot(1,3,1);
plt.imshow(imagey,cmap='gray')
plt.title('Ground Truth');
plt.axis('off')
# Step 1 generate projected pattern
g=signal.convolve2d(t,imagey, mode='same',boundary='wrap');
#The following line adds signal dependent noise
plt.subplot(1,3,2);
fe=cv2.resize(signal.convolve2d(ht,np.random.poisson(5*signal.convolve2d(t,imagey, mode='same',boundary='wrap'))\
                                                ,mode='same',boundary='wrap'),dsize=(28,28))
fe=fe/np.max(fe)
plt.imshow(fe, cmap='gray');
plt.title('Reconstructed image');
plt.axis('off')
plt.subplot(1,3,3);
plt.imshow(autoencoder.predict(np.reshape(fe,(1,28,28,1)))[0,:,:,0])
plt.gray()
plt.title('Denoised image');
plt.axis('off')
plt.savefig('muraCodedApertureReconRandom.pdf')
files.download('muraCodedApertureReconRandom.pdf')
```
